```python
#------------------------------------------------------------------------------------------------------------------------#
# Script responsável por ler um arquivo no formato 'csv'(Log.txt) e realizar o processo 'wrangling' para, posteriormente,#
# produzir um Dataframe. Ao final, gera um arquivo no formato 'csv' (log_csv)                                            #
# Autor: Ricardo AURÉLIO                                                                                                 #
#------------------------------------------------------------------------------------------------------------------------#
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
#
#---[Leitura e limpeza do arquivo csv de origem]
#
file_out = open("LogTmp.csv","w")
with open("Erro_FamFamiliaSmnPrd.txt",'r') as f:
    reg_in = 0
    reg_out = 0
    x = 0
    prim_vez = True
    listaCols = []
    listaCont = []

    for linha in f:
        listaCols = []
#----[Recupera identificação da aplicação na Log]        
        pos1 = linha.find("{")
        listaCols.append(linha[:pos1 - 1])
        idtLog = linha[:pos1 - 1]
#        print("Campo 1..: ",linha[:pos1 - 1])
#----[Recupera mensagem de inconsistência na Log]        
        pos2 = linha.find('}', pos1 + 2)
        pos3 = linha.find(':', pos1)
        msgLog = linha[pos3 + 2: pos2 - 1]
        msgLog = msgLog.replace(";", " ").replace('"','')
        print('MsgLog:', msgLog)
        listaCols.append(linha[pos3 + 2: pos2 - 1])
#        print("Campo 2..: ",linha[pos3 + 2: pos2 - 1])
#----[Recupera campos na Log] 
        posCampos = linha.find('{', pos2) + 1
        restoLinha = linha[posCampos:]
        #print('Resto linha:', restoLinha)
        listaTratar = ['seguradoEndereco', 'numeroLogradouro', 'complementoEndereco', 'seguradoBairro']
        for el in listaTratar:
            pi = restoLinha.find(el)
            pf = restoLinha.find(":", pi) + 2
            sci = restoLinha.find('"', pf)
            strSuja = restoLinha[pf: sci]
            strLimpa = strSuja.replace(",", "")
            restoLinha = restoLinha.replace(strSuja, strLimpa)

        tp = restoLinha.replace('","', '";"').replace('"', '').replace("'", "").replace("{", "").replace("}", "").split(";")
        camposRestantes = ""
        for el in tp:
            listaCols.append(str(el)[:str(el).find(":")])
            listaCont.append(str(el)[str(el).find(":")+1:len(str(el))])
            camposRestantes += str(el)[str(el).find(":")+1:len(str(el))] +";"
            
        regSaida = idtLog + ";" + msgLog + ";" + camposRestantes[:camposRestantes.rfind(";")]
        print('RegSaída:', regSaida.rstrip())
        file_out.write(regSaida.strip() + '\n')
        reg_out += 1
              
file_out.close
print(" Linhas de entrada: ", reg_in)
print("Linhas processadas: ", reg_out)
```
